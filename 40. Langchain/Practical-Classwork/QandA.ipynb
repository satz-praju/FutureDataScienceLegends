```python
import pinecone
from langchain_pinecone import PineconeVectorStore
from pinecone import ServerlessSpec
```

```python
pc = pinecone.Pinecone()
```

```python
index_name = "churchill-speech"
if not pc.has_index(index_name):
    pc.create_index(
        name=index_name,
        dimension=1536,
        metric="cosine",
        spec=ServerlessSpec(
            cloud='aws',
            region = 'us-east-1'
        )
            )
```

```python
vector_store = PineconeVectorStore.from_documents(texts, embeddings,index_name = index_name )
```

```python
vector_store
```

```python
query = "where should we fight?"
result = vector_store.similarity_search(query, k= 10 )
```

```python
result
```

```python
from langchain_pinecone import PineconeRerank

# Initialize the Pinecone Rerank with the desired model
reranker = PineconeRerank(model="bge-reranker-v2-m3", top_n=5)
```

```python
reranked_docs = reranker.compress_documents(result, query)
```

```python
reranked_docs
```

```python
from langchain.chains import RetrievalQA
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
```

```python
from re import search
retreiver = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 5})
chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retreiver,
                                    return_source_documents=True)
```

```python
query = "where should we fight?"
```

```python
chain.invoke(query)
```

```python
from langchain.retrievers import ContextualCompressionRetriever

reranking_retriever = ContextualCompressionRetriever(
    base_retriever=retreiver,
    base_compressor=reranker
)
```

```python
from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=reranking_retriever,
    return_source_documents=True
)
```

```python
query = "where should we fight?  give me the response in the bullet points?"
chain.invoke(query)
```
